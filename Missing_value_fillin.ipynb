```python
# find the column names for Macroeonomic Data needs revision, 
# we do not have 2025Q2 available yet

import re

# Macroeconomic variable prefixes
macros = ["GDP", "GDPReal", "Unemployment", "InterestRate", "IndustrialProd", "Inflation"]

def shift_quarter(col):
    # check if column starts with one of the macro prefixes
    if any(col.startswith(m) for m in macros):
        m = re.match(r"(.+)_(\d{4})Q(\d)", col)
        if m:
            var, year, q = m.groups()
            year = int(year)
            q = int(q)

            # shift back one quarter
            if q == 1:
                q = 4
                year -= 1
            else:
                q -= 1

            return f"{var}_{year}Q{q}"
    return col  # return unchanged if not macro

# update only matching columns
df.columns = [shift_quarter(c) for c in df.columns]
```

```python
df.columns.tolist()
```

```python
# Count missing values and total per column
missing_summary = (
    df.isna()
       .sum()
       .to_frame("MissingCount")
       .assign(Total=df.shape[0])
       .assign(MissingPct=lambda d: d["MissingCount"] / d["Total"] * 100)
)

# Show a preview
missing_summary.head(30)
```

```python
# In order to filling missing values, 
# we need to understand the distribution of 'Market Value'
# we fill in missing values with median of each tranche, based on the same sector
df['Market Value'] = (
    df['Market Value']
    .replace('[\$,]', '', regex=True)   # remove $ and commas
    .astype(float)                      # convert to float
)

df['Market Value'].describe()
```

```python
# Define bins and labels
bins = [
    0, 0.05e6, 0.1e6, 0.15e6, 0.2e6,      # split <0.2m
    0.4e6, 0.6e6, 0.8e6, 1e6,             # keep sub-1m splits
    2e6, 3.5e6, 5e6,                      # split 1â€“5m
    10e6, 100e6, 500e6, 1e9, float('inf') # collapse big end
]

labels = [
    '<0.05m', '0.05â€“0.1m', '0.1â€“0.15m', '0.15â€“0.2m',
    '0.2â€“0.4m', '0.4â€“0.6m', '0.6â€“0.8m', '0.8â€“1.0m',
    '1â€“2m', '2â€“3.5m', '3.5â€“5m',
    '5â€“10m', '10â€“100m', '100â€“500m', '500mâ€“1b', '1b+'
]

df['MarketCap_Tranche'] = pd.cut(df['Market Value'], bins=bins, labels=labels, include_lowest=True, right=False)
df['MarketCap_Tranche'].value_counts().sort_index()
```

```python
fundamental_prefixes = [
    "CapitalExpenditure", "CashAndSTInvestments", "CashFromOps", "CostOfRevenue",
    "CurrentAssets", "CurrentLiabilities", "EPS", "IncomeTaxExpense", "InterestExpense",
    "LongTermDebt", "NetIncome", "OperatingIncome", "OtherOperatingExpense",
    "Revenue", "TotalAssets", "TotalDebt", "TotalEquity", "TotalLiabilities"
]
target_cols = [c for c in df.columns if any(c.startswith(p + "_") for p in fundamental_prefixes)]
target_cols
```

```python
# Sector + Tranche
for col in target_cols:
    df[col] = df.groupby(['Sector', 'MarketCap_Tranche'])[col].transform(lambda s: s.fillna(s.median()))


# Remaining NaN counts for target columns
remaining_nans = {col: int(df[col].isna().sum()) for col in target_cols}

print("\nRemaining NaNs in target columns (should be 0 after fallback):")
remaining_nans
```

```python
df = df.dropna()
df.shape
```
