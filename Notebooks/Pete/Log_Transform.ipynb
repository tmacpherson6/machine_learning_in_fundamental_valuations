## Log Transformation

Markets tend to exhibit a log-normal distribution across companies for many aspects of financial data.

```python
import sys
PATH_TO_MODULES = '../..'
sys.path.insert(0, PATH_TO_MODULES)

from helpers import get_quarters, log_transform

# Data Manipulation and Math/Stats Libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Import data
path = PATH_TO_MODULES + '/datasets/'
filename = 'X_train_filled_KPIs_QoQ.csv'
dataset = pd.read_csv(path + filename)
```

## Distribution of Companies by log(Market Value)

```python
plt.hist(dataset['Market Value'].apply(np.log), bins=50);
```

```python
quarters = get_quarters(dataset)
print(quarters)
```

```python
# Checking zero values that cause a problem for the log transform
var = 'TotalDebt'
for quarter in quarters:
    rows_where_zero = dataset[var + quarter][dataset[var + quarter] == 0].shape[0]
    print(f'{rows_where_zero} rows with value zero in {var + quarter}')
```

```python
log_vars = ['Revenue',
            # 'NetIncome',
            'CurrentAssets',
            'CurrentLiabilities',
            'TotalAssets',
            'TotalLiabilities',
            'TotalEquity',
            # 'Total Debt'
            'CashFromOps'
           ]
```

**Check to see if a log transformation helps with columns that inform KPIs.**

```python
for var in log_vars:
    for quarter in quarters:
        print(var, quarter[1:])
        plt.hist(dataset[var + quarter].apply(np.log), bins=50)
        plt.title(f'Distribution of log({var}) for ' + quarter[1:])
        plt.show()
```
