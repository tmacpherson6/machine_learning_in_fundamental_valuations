# KPI Exploration I

This notebook explores different transformations on the KPIs and visualizes their distributions using boxplots.

```python
import sys
PATH_TO_MODULES = '../../..'
sys.path.insert(0, PATH_TO_MODULES)

from helpers import *
from unsupervised_helpers import *
```

```python
import seaborn as sns
```

```python
# Import merged dataset
path = PATH_TO_MODULES + '/datasets/'
input_filename = 'X_train_filled_KPIs_QoQ.csv'
dataset = pd.read_csv(path + input_filename)
if 'Unnamed: 0' in dataset.columns:
    dataset.drop(columns=['Unnamed: 0'], inplace=True)
print(f'There are {dataset.shape[0]} rows and {dataset.shape[1]} columns in the dataset.')
```

```python
KPIs = [col for col in dataset.columns if 'KPI' in col]
dataset[KPIs].describe()
```

```python
KPIs_2025Q1 = [col for col in dataset[KPIs].columns if '2025Q1' in col]
dataset[KPIs_2025Q1].describe()
```

```python
KPI_df = dataset[KPIs_2025Q1]
KPI_df.index = dataset['Ticker']
```

```python
scaler_qt = QuantileTransformer()
full_KPI_scaled_df = pd.DataFrame(
    scaler_qt.fit_transform(full_KPI_df),
    columns=full_KPI_df.columns
)
full_KPI_scaled_df.describe()
```

```python
boxplot_df = full_KPI_scaled_df.melt(value_name='ratio')
sns.boxplot(
    boxplot_df,
    x="ratio", 
    y="variable"
)
```

```python
interquartile_index = set(KPI_df.index)
for col in KPI_df.columns:
    above_first_quartile = KPI_df[col] >= KPI_df[col].describe().loc['25%']
    below_third_quartile = KPI_df[col] <= KPI_df[col].describe().loc['75%']
    interquartile = KPI_df[col][(above_first_quartile) & (below_third_quartile)]
    count = interquartile.count()
    print(col, f': {count} companies within interquartile range')
    interquartile_index = interquartile_index & set(interquartile.index)
print(f'{len(interquartile_index)} companies within the interquartile range for all KPIs')
```

```python
iq_df = KPI_df.loc[list(interquartile_index)]
```

```python
boxplot_df = iq_df.melt(value_name='ratio')
sns.boxplot(
    boxplot_df,
    x="ratio", 
    y="variable"
)
```

```python
scaler1 = QuantileTransformer(n_quantiles=100, output_distribution='uniform')
iq_scaled_df = pd.DataFrame(scaler1.fit_transform(iq_df), columns=iq_df.columns)
iq_scaled_df.describe()
```

```python
boxplot_df = iq_scaled_df.melt(value_name='ratio')
sns.boxplot(
    boxplot_df,
    x="ratio", 
    y="variable"
)
```

```python
scaler1 = QuantileTransformer(n_quantiles=100, output_distribution='uniform')
KPI_scaled_df = pd.DataFrame(scaler1.fit_transform(KPI_df), columns=iq_df.columns)
KPI_scaled_df.describe()
```

```python
boxplot_df = KPI_scaled_df.melt(value_name='ratio')
sns.boxplot(
    boxplot_df,
    x="ratio", 
    y="variable"
)
```

```python
scaler2 = MinMaxScaler(feature_range=(0, 100))
iq_minmax_df = pd.DataFrame(
    scaler2.fit_transform(iq_df), columns=iq_df.columns
)
iq_minmax_df.describe()
```

```python
boxplot_df = iq_minmax_df.melt(value_name='ratio')
sns.boxplot(
    boxplot_df,
    x="ratio", 
    y="variable"
)
```
