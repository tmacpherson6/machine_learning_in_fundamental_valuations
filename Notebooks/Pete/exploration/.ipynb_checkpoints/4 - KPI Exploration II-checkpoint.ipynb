```python
# Make new dataframe keeping only KPI's and the target (Net Income)
KPIs, covariates = list(get_KPI(dataset).columns), []
for KPI in KPIs:
    if '2024Q4' in KPI:
        covariates.append(KPI)
regressand = ['Revenue_2025Q1']
analysis_df = dataset[covariates + regressand]
```


## Linear Regression

### Predict EPS_2025Q2 using KPI's from 2025Q1

Attempting a linear regression at this stage will allow us to examine the residuals to see if there are any patterns.  The presence of a pattern might indicate a need to transform either the regressand, the covariate, or both, to obtain a linear relationship.

```python
y = analysis_df[['Revenue_2025Q1']]
X = analysis_df.drop(columns=['Revenue_2025Q1'])
model = sm.OLS(y, sm.add_constant(X))
results = model.fit()
print(results.summary())
```

R-squared of 0.047 indicates that a simple linear model using KPI's explains very little about the observed EPS for the next quarter.  Let's see if we can find any patterns in these residuals.

**Next steps:**
 - examine residuals of each covariate vs. the regressand for patterns
 - if a pattern is found, apply an appropriate transformation to obtain a linear relationship


### Plot of residuals for multivariate regression

```python
fig = plt.figure()
ax = fig.subplots()
ax.scatter(y, results.resid)
ax.set_xlabel(f'{regressand[0]} (observed)')
ax.set_ylabel('Residual from predictions')
plt.show()
```

The plot shows a linear relationship between the residuals (the noise) and the regressand: as the observed value of the regressand increases, the residual from the predication also increases.  Linear models assume zero covariance between residuals and the regressand, so we can see that a linear model is not a good fit for this dataset.  The residuals and the regressand share a homoscedastic relationship, which tends to exaggerate the coefficient of determination (R-squared).


### Plot of each residual distribution for univariate regressions


```python
for column in X:
    plot_residuals(X[column], y)
```

These plots indicate that rthe esidual distribution for each individual KPI often has some kind of relationship with the regressand, which violates assumptions of the linear model.  We can also see that none of the residuals for KPIs are normally distributed.

For these reasons, a simple linear regression does not appear to be a good choice.
