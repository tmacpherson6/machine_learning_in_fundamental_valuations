```python


def plot_categorical_dist(feature):
    """Plot a distribution (histogram) for a categorical feature."""
    fig = px.histogram(
        dataset,
        x=feature, y='Ticker', histfunc='count',
        title='Distribution of Companies by ' + feature,
        width=600, height=400
    )
    fig.update_layout(title_pad_l=120)
    fig.show()
    return None

    
```

Let's examine company financials in the context of different categorical variables.


### Sector

```python
quarter = '2025Q2'
columns = [f'Revenue_{quarter}',
           f'OperatingIncome_{quarter}',
           f'NetIncome_{quarter}',
           f'CashFromOps_{quarter}',
           f'CashAndSTInvestments_{quarter}',
           f'LongTermDebt_{quarter}',
           f'TotalAssets_{quarter}',
           f'TotalLiabilities_{quarter}',
           f'EPS_{quarter}'
          ]

fig = px.scatter_matrix(
    dataset,
    dimensions=columns,
    color='Sector',
    width=1280, height=720
)
fig.show()
```
