## Log Transformation

Markets tend to exhibit a log-normal distribution across companies for many aspects of financial data.  This notebook investigates how a logarithmic transformation of the base variables improves their distribution.

```python
import sys
PATH_TO_MODULES = '../../..'
sys.path.insert(0, PATH_TO_MODULES)

from helpers import get_quarters, log_transform

# Data Manipulation and Math/Stats Libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
import plotly.express as px
import plotly.graph_objects as go
```

```python
from matplotlib.gridspec import GridSpec
```

```python
# Import data
path = PATH_TO_MODULES + '/datasets/'
filename = 'X_train_filled_KPIs_QoQ.csv'
dataset = pd.read_csv(path + filename)
print(f'There are {dataset.shape[0]} rows and {dataset.shape[1]} columns in the data')
```

## Distribution of Companies by Market Value

```python
fig = plt.figure(figsize=(10, 2))
gs = GridSpec(1, 2, wspace=0.5)
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])
# Raw data
ax1.hist(dataset['Market Value'] / 10**6, bins=50)
ax1.set_title('Distribution of Market Value')
ax1.set_ylabel('Count of Companies')
ax1.set_xlabel('bins of Market Value (billions USD)')
# Log-transformed data
ax2.hist((dataset['Market Value'] * 1000 + 2).apply(np.log10), bins = 50)
ax2.set_title('Distribution of log_10(Market Value)')
ax2.set_ylabel('Count of Companies')
ax2.set_xlabel('bins of log_10(Market Value)')
plt.show()
```

We can see that before transformation, nearly all companies fall in a very narrow range near the median value, while a few extremely large companies (the 'Mega Cap' group) accounts for most of the variance in the data.

After transformation, the distribution of market value looks almost bell-shaped, and we can more easily distinguish nuances between the companies near the median.

```python
quarters = get_quarters(dataset)
print(quarters)
```

```python
# Checking zero values that cause a problem for the log transform
var = 'TotalEquity'
for quarter in quarters:
    rows_where_zero = dataset[var + quarter][dataset[var + quarter] <= 0].shape[0]
    print(f'{rows_where_zero} rows with value <= 0 in {var + quarter}')
```

```python
KPI_vars = ['Revenue',
            #'NetIncome',
            'CurrentAssets',
            'CurrentLiabilities',
            'TotalAssets',
            'TotalLiabilities',
            #'TotalEquity',
            'TotalDebt',
            #'CashFromOps'
           ]
for var in KPI_vars:
    for quarter in quarters:
        rows_where_zero = dataset[var + quarter][dataset[var + quarter] <= 0].shape[0]
        print(f'{rows_where_zero} rows with value <= 0 in {var + quarter}')
```

**Check to see if a log transformation helps with columns that inform KPIs.**

The plots below show that, for variables where values are consistently non-negative, a log transformation improves the distribution considerably.

*Note: Many companies had negative values for Net Income, Total Equity, and CashFromOps, so they were excluded from this analysis.*

```python
for var in KPI_vars:
    for quarter in quarters:
        print(var, quarter[1:])
        plt.hist(dataset[var + quarter].apply(np.log), bins=50)
        plt.title(f'Distribution of log({var}) for ' + quarter[1:])
        plt.show()
```
